1.Generate a normally distributed data, with mean=3 and standard deviation=2, for 10,000 observations and 
Plot the normal distribution overlaid on top of a histogram in matplotlib so you get the following result:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
mu,sigma = 3,2
population= np.random.normal(mu,sigma,10000)
population.sort()
```

```python
 #plotting histograms
plt.hist(population,bins=30,density=True)
# plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
#                np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
#          linewidth=2, color='r')

p = norm.pdf(population,mu,sigma) #Using Guassian probability density function to produce normal distribution curve using
                            #provided mu and sigma
plt.plot(population,p,color="red") #plotting normal curve
plt.title("Hypothesis Testing with Python")
conf_int_s=norm.interval(0.95,mu,sigma)
```
