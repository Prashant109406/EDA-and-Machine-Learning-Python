### Hypothesis 1:People of california have the highest death count due to drugs use with std cases

```python
count1=Data_Merge.groupby(['Year','State','Gender']).sum().Deaths
count2=Data_Merge.groupby(['Year','State','Gender']).count().Deaths

count=count1/count2

hypo1=count.sort_values(ascending = False) 
hypo1=hypo1.to_frame()
```

```python
hypo1=hypo1.groupby(['State']).sum().reset_index()
hypo1
```

```python
figure(num=None, figsize=(10, 10), dpi=80, facecolor='w', edgecolor='k')
sns.set(font_scale=1.1)
sns.barplot(x='State',y='Deaths',data=hypo1).set_title("Death count Statewise")
plt.ticklabel_format(style='plain', axis='y',useOffset=False)
plt.xticks(rotation=90)
plt.show()
```

As per the above analysis we can validate Hypothesis 1 where People of california have the highest death count due to drugs use with std casesÂ¶

### Hypothesis 2 : males with std cases have high mortality to die thorugh drug induced death

```python
count1=Data_Merge.groupby(['Year','State','Gender',]).sum().Deaths
count2=Data_Merge.groupby(['Year','State','Gender']).count().Deaths
count=count1/count2
print(count)
```

```python
hypo2=count.to_frame()
hypo2
```

```python
hypo2=hypo2.groupby(['Gender']).sum().reset_index()
hypo2['Deaths']=hypo2['Deaths'].astype(int)
```

```python
hypo2
```

```python
sns.set(font_scale=1.1)
sns.barplot(x='Gender',y='Deaths',data=hypo2).set_title("Death count Male vs Female")
plt.ticklabel_format(style='plain', axis='y',useOffset=False)
plt.show()
```

From the above analysis we can invalidate Hypothesis 2. Females with std cases have high death count.

### Question : Do people with Chlamydia cases have the highest number of cases and have the highest mortality rate due to drug overdose in 2013?

```python
#making dataframe for only Chlamydia cases
df4=Data_Merge.loc[Data_Merge['Disease'] == 'Chlamydia']
```

```python
#drugs deaths
count2=df4.groupby(['Year','Disease']).sum().Deaths
```

```python
figure(num=None, figsize=(10, 10), dpi=80, facecolor='w', edgecolor='k')

count2.plot.pie(shadow=False,startangle=0,autopct='%1.1f%%',fontsize=15)
plt.xlabel("Death Count \n(%)",fontsize=15)
plt.title("Drugs death yearwise in %")
plt.ylabel("")
```

2013 accounting to 6.6% of death cases stands at 2nd position whereas 2014 has the highest count

-----------------------------

```python
#std cases
count3=df4.groupby(['Year','Disease']).sum().STD_Cases
```

```python
c=count3.sort_values(ascending = False) 
figure(num=None, figsize=(10, 10), dpi=80, facecolor='w', edgecolor='k')

c.plot.pie(shadow=False,startangle=0,autopct='%1.1f%%',fontsize=15)
plt.xlabel("Death Count \n(%)",fontsize=15)
plt.title("STD cases yearwise in %")
plt.ylabel("")
```

2013 accounting to 8.6% of STD cases stands at 2nd position whereas 2014 has the highest cases

No people with Chlamydia cases do not have the highest cases and mortality rate due to drug overdose in 2013 but they have the highest STD cases and drugs death in 2014.
